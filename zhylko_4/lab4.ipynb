```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as mcoll
import matplotlib.colors as mcolors
from matplotlib.widgets import Button
import json as js

TOLERANCE = 0.15

def dist(point1, point2):
    return np.sqrt(np.power(point1[0] - point2[0], 2) + np.power(point1[1] - point2[1], 2))

class _Button_callback(object):
    def __init__(self, scenes):
        self.i = 0
        self.scenes = scenes
        self.adding_points = False
        self.added_points = []
        self.adding_lines = False
        self.added_lines = []
        self.adding_rects = False
        self.added_rects = []

    def set_axes(self, ax):
        self.ax = ax
        
    def next(self, event):
        self.i = (self.i + 1) % len(self.scenes)
        self.draw(autoscaling = True)

    def prev(self, event):
        self.i = (self.i - 1) % len(self.scenes)
        self.draw(autoscaling = True)
        
    def add_point(self, event):
        self.adding_points = not self.adding_points
        self.new_line_point = None
        if self.adding_points:
            self.adding_lines = False
            self.adding_rects = False
            self.added_points.append(PointsCollection([]))
         
    def add_line(self, event):   
        self.adding_lines = not self.adding_lines
        self.new_line_point = None
        if self.adding_lines:
            self.adding_points = False
            self.adding_rects = False
            self.added_lines.append(LinesCollection([]))

    def add_rect(self, event):
        self.adding_rects = not self.adding_rects
        self.new_line_point = None
        if self.adding_rects:
            self.adding_points = False
            self.adding_lines = False
            self.new_rect()
    
    def new_rect(self):
        self.added_rects.append(LinesCollection([]))
        self.rect_points = []
    
    def on_click(self, event):
        if event.inaxes != self.ax:
            return
        new_point = (event.xdata, event.ydata)
        if self.adding_points:
            self.added_points[-1].add_points([new_point])
            self.draw(autoscaling = False)
        elif self.adding_lines:
            if self.new_line_point is not None:
                self.added_lines[-1].add([self.new_line_point, new_point])
                self.new_line_point = None
                self.draw(autoscaling = False)
            else:
                self.new_line_point = new_point
        elif self.adding_rects:
            if len(self.rect_points) == 0:
                self.rect_points.append(new_point)
            elif len(self.rect_points) == 1:
                self.added_rects[-1].add([self.rect_points[-1], new_point])
                self.rect_points.append(new_point)
                self.draw(autoscaling = False)
            elif len(self.rect_points) > 1:
                if dist(self.rect_points[0], new_point) < (np.mean([self.ax.get_xlim(), self.ax.get_ylim()])*TOLERANCE):
                    self.added_rects[-1].add([self.rect_points[-1], self.rect_points[0]])
                    self.new_rect()
                else:    
                    self.added_rects[-1].add([self.rect_points[-1], new_point])
                    self.rect_points.append(new_point)
                self.draw(autoscaling = False)
        
    def draw(self, autoscaling = True):
        if not autoscaling:
            xlim = self.ax.get_xlim()
            ylim = self.ax.get_ylim()
        self.ax.clear()
        for collection in (self.scenes[self.i].points + self.added_points):
            if len(collection.points) > 0:
                self.ax.scatter(*zip(*(np.array(collection.points))), **collection.kwargs)
        for collection in (self.scenes[self.i].lines + self.added_lines + self.added_rects):
            self.ax.add_collection(collection.get_collection())
        self.ax.autoscale(autoscaling)
        if not autoscaling:
            self.ax.set_xlim(xlim)
            self.ax.set_ylim(ylim)
        plt.draw()

```

```python
class Scene:
    def __init__(self, points=[], lines=[]):
        self.points=points
        self.lines=lines

class PointsCollection:
    def __init__(self, points, **kwargs):
        self.points = points
        self.kwargs = kwargs
    
    def add_points(self, points):
        self.points = self.points + points

class LinesCollection:
    def __init__(self, lines, **kwargs):
        self.lines = lines
        self.kwargs = kwargs
        
    def add(self, line):
        self.lines.append(line)
        
    def get_collection(self):
        return mcoll.LineCollection(self.lines, **self.kwargs)
    
class Plot:
    def __init__(self, scenes = [Scene()], json = None):
        if json is None:
            self.scenes = scenes
        else:
            self.scenes = [Scene([PointsCollection(pointsCol) for pointsCol in scene["points"]], 
                                 [LinesCollection(linesCol) for linesCol in scene["lines"]]) 
                           for scene in js.loads(json)]
        
    def __configure_buttons(self):
        plt.subplots_adjust(bottom=0.2)
        ax_prev = plt.axes([0.6, 0.05, 0.15, 0.075])
        ax_next = plt.axes([0.76, 0.05, 0.15, 0.075])
        ax_add_point = plt.axes([0.44, 0.05, 0.15, 0.075])
        ax_add_line = plt.axes([0.28, 0.05, 0.15, 0.075])
        ax_add_rect = plt.axes([0.12, 0.05, 0.15, 0.075])
        b_next = Button(ax_next, 'Następny')
        b_next.on_clicked(self.callback.next)
        b_prev = Button(ax_prev, 'Poprzedni')
        b_prev.on_clicked(self.callback.prev)
        b_add_point = Button(ax_add_point, 'Dodaj punkt')
        b_add_point.on_clicked(self.callback.add_point)
        b_add_line = Button(ax_add_line, 'Dodaj linię')
        b_add_line.on_clicked(self.callback.add_line)
        b_add_rect = Button(ax_add_rect, 'Dodaj figurę')
        b_add_rect.on_clicked(self.callback.add_rect)
        return [b_prev, b_next, b_add_point, b_add_line, b_add_rect]
    
    def add_scene(self, scene):
        self.scenes.append(scene)
    
    def add_scenes(self, scenes):
        self.scenes = self.scenes + scenes

    def toJson(self):
        return js.dumps([{"points": [np.array(pointCol.points).tolist() for pointCol in scene.points], 
                          "lines":[linesCol.lines for linesCol in scene.lines]} 
                         for scene in self.scenes])    
     
    def get_added_points(self):
        if self.callback:
            return self.callback.added_points
        else:
            return None
  
    def get_added_lines(self):
        if self.callback:
            return self.callback.added_lines
        else:
            return None
        
    def get_added_figure(self):
        if self.callback:
            return self.callback.added_rects
        else:
            return None
    
    def get_added_elements(self):
        if self.callback:
            return Scene(self.callback.added_points, self.callback.added_lines+self.callback.added_rects)
        else:
            return None
    
    def draw(self):
        plt.close()
        fig = plt.figure()
        self.callback = _Button_callback(self.scenes)
        self.widgets = self.__configure_buttons()
        ax = plt.axes(autoscale_on = False)
        self.callback.set_axes(ax)
        fig.canvas.mpl_connect('button_press_event', self.callback.on_click)
        plt.show()
        self.callback.draw()
        
```

```python
def det(a,b,c):
    return a[0]*b[1] + a[1]*c[0] + b[0]*c[1] - b[1]*c[0] - a[1]*b[0] - a[0]*c[1]
```

```python
def get_orientation(a,b,c, eps = 1e-6):
    orient = det(a,b,c)
    
    if(orient > eps):
        return 1
    elif(orient < -eps):
        return -1
    else:
        return 0
```

```python
from enum import Enum

class VertexType(Enum):
    start = 'start'
    end = 'end'
    merge = 'merge'
    split = 'split'
    regular = 'regular'
    
    @staticmethod
    def type_of_vertex(before, current, after):
        orient = get_orientation(before, current, after)
        higher_than_before = before[1] < current[1]
        higher_than_after = after[1] < current[1]
        
        if higher_than_before and higher_than_after:
            return VertexType.start if orient == 1 else VertexType.split
        if not (higher_than_before or higher_than_after):
            return VertexType.end if orient == 1 else VertexType.merge
        return VertexType.regular
```

```python
def is_y_monotonic(points):
    points_to_types = []
    result = True
    for i in range(len(points)):
        
        before_index = i-1
        after_index = (i+1)%len(points)
        
        vertex_type = VertexType.type_of_vertex(*points[[before_index, i, after_index]])
        
        points_to_types.append((points[i], vertex_type))
        
        if(vertex_type in [VertexType.merge, VertexType.split]):
            result = False
    return result, points_to_types
```

```python
def find_extremum(points, comparator):
    p_m = points[0]
    p_m_index = 0
    
    for idx in range(1, len(points)):
        p_m, p_m_index = comparator(p_m, p_m_index, points[idx], idx)
    return p_m, p_m_index
        
def max_comparator(a, idxa, b, idxb):
    return (a, idxa) if (a[1] > b[1]) or (a[1] == b[1] and a[0] > b[0]) else (b, idxb)

def min_comparator(a, idxa, b, idxb):
    return (a, idxa) if (a[1] < b[1]) or (a[1] == b[1] and a[0] < b[0]) else (b, idxb)

def get_chain(points, start, end):
    chain = []
    n = len(points)
    
    while(start != end):
        chain.append(start)
        start = (start + 1) % n
    chain.append(end)
    return chain

def join_all(points, triangulation, p_idx, vertexes, poligon_edges):
    while(len(vertexes) > 0):
        v_idx = vertexes.pop()
        if((p_idx, v_idx) not in poligon_edges):
            triangulation.append([None, p_idx, v_idx]) # we don't have information about 3 vertex of the triangel

def join_if(points, triangulation, p_idx, stack, orientation): # , poligon_edges
    stack_top_idx = stack.pop()
    
    while len(stack) > 0:
        v_idx = stack[-1]
        
        if get_orientation(*points[[stack_top_idx, p_idx, v_idx]]) == orientation: # and (p_idx, v_idx) not in poligon_edges
            triangulation.append([stack_top_idx, p_idx, v_idx])
        else:
            break
        stack_top_idx = stack.pop()
    stack.append(stack_top_idx)
    stack.append(p_idx)

def triangulate(points):
    is_monotonic, _ = is_y_monotonic(points)
    if not is_monotonic:
        print('wielokąt nie jest y monotoniczny')
        return None, None 
    
    max_v, max_index = find_extremum(points, max_comparator)
    min_v, min_index = find_extremum(points, min_comparator)
    
    left_chain = set(get_chain(points, max_index, min_index))
    
    right_chain = set(get_chain(points, min_index, max_index))
    
    vertexes = [i for i in range(len(points))]
    
    vertexes.sort(key=lambda x: points[x][1], reverse=True)
    
    poligon_edges = set()
    
    n = len(points)
    
    for i in range(n):
        poligon_edges.add(((i-1)%n, i))
        poligon_edges.add((i, (i-1)%n))
        
    stack = [vertexes[0], vertexes[1]]
    
    visualization_steps = []
    triangulation = []
    
    for p_idx in vertexes[2:]:
        visualization_steps.append([stack.copy(), p_idx, triangulation.copy()])
        stack_top_idx = stack[-1]
        
        if stack_top_idx in left_chain and p_idx in left_chain:
            join_if(points, triangulation, p_idx, stack, 1)
        elif stack_top_idx in right_chain and p_idx in right_chain:
            join_if(points, triangulation, p_idx, stack, -1)
        else:
            join_all(points, triangulation, p_idx, stack, poligon_edges)
            stack = [stack_top_idx, p_idx]
    visualization_steps.append([stack.copy(), vertexes[-1], triangulation.copy()])
    return triangulation, visualization_steps
```

```python
def visualize_poligon_with_triangulation(points, triangulation):
    poligon = []
    
    for i in range(len(points)):
        poligon.append(points[[i-1, i]])
    
    #print(triangulation)
    triangulation_edges = []
    
    for i in triangulation:
        triangulation_edges.append(points[i[1:]])
    
    scene = Scene([], [LinesCollection(poligon), LinesCollection(triangulation_edges, color='red')])
    plot = Plot([scene])
    return plot
```

```python
def visualize_algorithm_steps(points, steps):
    poligon = []
    
    for i in range(len(points)):
        poligon.append(points[[i-1, i]])
    
    scenes = []
    
    for i in steps:
        triangulation_edges = []
    
        for j in i[2]:
            triangulation_edges.append(points[j[1:]])
        
        scenes.append(Scene([PointsCollection(points[i[0]], color='red'), PointsCollection([points[i[1]]], color='magenta')], [LinesCollection(triangulation_edges, color='red'), LinesCollection(poligon, linestyle='dashed')]))
    
    plot = Plot(scenes)
    return plot
```

```python
def visualize_poligon_with_colors(points, colors):
    poligon = []
    
    for i in range(len(points)):
        poligon.append(points[[i-1, i]])
    
    start = []
    end = []
    split = []
    merge = []
    regular = []
    
    for p, c in colors:
        if c == VertexType.start:
            start.append(p)
        if c == VertexType.end:
            end.append(p)
        if c == VertexType.split:
            split.append(p)
        if c == VertexType.merge:
            merge.append(p)
        if c == VertexType.regular:
            regular.append(p)
    
    scene = Scene([PointsCollection(start, color = 'green'), PointsCollection(end, color='red'), \
                   PointsCollection(split, color='blue'), PointsCollection(merge, color='magenta'),\
                  PointsCollection(regular, color='black')], [LinesCollection(poligon)])
    plot = Plot([scene])
    return plot
```

```python
def fix_orientation(points, triangle):
    if get_orientation(*points[triangle]) != 1:
        triangle[1], triangle[2] = triangle[2], triangle[1]
    
def determin_triangels(points, triangulation):
    n = len(points)
    
    G = [set() for i in range(n)]
    
    for i in range(n):
        G[(i-1)%n].add(i)
        G[i].add((i-1)%n)
    
    for _, a, b in triangulation:
        G[a].add(b)
        G[b].add(a)
    
    for i in range(len(triangulation)):
        if triangulation[i][0] is not None:
            fix_orientation(points, triangulation[i])
        else:
            # find vertex neighbouring with both verteces
            for j in G[triangulation[i][1]]:
                if j in G[triangulation[i][2]]:
                    triangulation[i][0] = j
            
            fix_orientation(points, triangulation[i])
```

```python
def write_triangulation(file_name, triangulation, points):
    f = open(file_name + '.abc', 'wb')
    import pickle
    
    triangulation_points = []
    
    for i in triangulation:
        triangulation_points.append(points[i])
    
    pickle.dump(triangulation_points, f)
    
    f.close()
```

```python
def visualize_from_file(filename):
    f = open(filename, 'rb')
    import pickle
    triangulation = pickle.load(f)
    
    edges = []
    for a,b,c in triangulation:
        edges.append([a,b])
        edges.append([b,c])
        edges.append([a,c])
    
    scene = Scene([],[LinesCollection(edges, color='darkorange')])
    plot = Plot([scene])
    return plot
```
